```python
rutera = 'C:/Users/felip/OneDrive/Documentos/TUPED/3erAño/Visualizacion/data/casos_covid19_2024.csv'
df = pd.read_csv(rutera, sep = ";")
```

```python
#elimino columna 'comuna'
df = df.drop(columns = ['comuna', 'tipo_contagio',])

```

```python
#descarto la hora no registrada y lo convierto a formato fecha(date) sin time
df['fecha_apertura_snvs'] = pd.to_datetime(df['fecha_apertura_snvs'], format='%d%b%Y:%H:%M:%S.%f').dt.date
df['fecha_toma_muestra'] = pd.to_datetime(df['fecha_toma_muestra'], format='%d%b%Y:%H:%M:%S.%f', errors='coerce').dt.date
df['fecha_clasificacion'] = pd.to_datetime(df['fecha_clasificacion'], format='%d%b%Y:%H:%M:%S.%f').dt.date
df['fecha_fallecimiento'] = pd.to_datetime(df['fecha_fallecimiento'], format='%d%b%Y:%H:%M:%S.%f', errors='coerce').dt.date
df['fecha_alta'] = pd.to_datetime(df['fecha_alta'], format='%d%b%Y:%H:%M:%S.%f', errors='coerce').dt.date
```

```python
df.head()
```

```python
#en la columna 'fallecido' cambio el valor nan por 'no'
df['fallecido'] = df['fallecido'].fillna('no')
```

```python
df['edad'] = df['edad'].fillna(0).astype(int)
```

```python
df.head()
```

```python
df['provincia'].isnull().sum()
```

```python
# Seleccionar los valores no nulos de la columna 'provincia'
valores_no_nulos = df['provincia'].dropna().unique()

# Reemplazar los valores nulos por valores aleatorios de los no nulos
df['provincia'] = df['provincia'].apply(lambda x: np.random.choice(valores_no_nulos) if pd.isnull(x) else x)

# Verificar los cambios
print(df['provincia'].isnull().sum())  # Debería ser 0 si todos los valores nulos han sido reemplazados
```

```python
#renombro la variable numero de caso
df = df.rename(columns={'numero_de_caso': 'id_caso'})
#convierto a str
df['id_caso'] = df['id_caso'].astype(str)
```

```python
#Guardar en un csv
df.to_csv('C:/Users/felip/OneDrive/Documentos/TUPED/3erAño/Visualizacion/data/casos_covid19_2024_clean.csv', index = False)
```

```python
import pandas as pd
import statsmodels.api as sn 
```

```python
archi = pd.read_csv("C:/Users/usr/Documents/TUPED/prubeas/datasets/world-happiness-2016.csv")
```

```python
variables = archi.columns
print(variables)
```

```python
x = archi[['Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Generosity', 
           'Freedom', 'Trust (Government Corruption)', 'Family']]
y = archi['Happiness Score']
```

```python
X = sn.add_constant(x)
```

```python
#ajustar el modelo de regresion lineal multiple
modelo = sn.OLS(y, X)
resultado = modelo.fit()
print(resultado.summary())
```

```python
#crear graficos de dispersion multiple
import matplotlib.pyplot as plt
fig,axs = plt.subplots(len(X.columns), figsize = (8, 6*len(X.columns)))

#iterar sobre las variables independientes
for i, column in enumerate(X.columns):
    axs[i].scatter(X[column],y)
    axs[i].set_xlabel(column)
    axs[i].set_ylabel('Happiness Score')

#ajustar espacios entre gráficos
plt.tight_layout()

#mostrar gradicos de dispersion
plt.show()
```
